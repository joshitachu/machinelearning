```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Basisstijl voor plots
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["axes.grid"] = True

```

```python

# Laad de dataset
df = pd.read_csv("ML-sleep_health_lifestyle_dataset.csv")

# Eerste rijen bekijken
df.head()


```

```python
# Structuur en datatypes
print("Informatie over de dataset:")
df.info()

print("\nBeschrijving van numerieke kolommen:")
display(df.describe())

print("\nBeschrijving van alle kolommen:")
display(df.describe(include='all'))

print("\nAantal ontbrekende waarden per kolom:")
print(df.isnull().sum())

```

```python
# Numerieke kolommen bepalen
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("Numerieke kolommen:", list(num_cols))

# Histogrammen van numerieke variabelen
df[num_cols].hist(figsize=(15, 12), bins=20)
plt.suptitle("Histogram", fontsize=16)
plt.show()

# Verdeling van de targetvariabele 'Sleep Disorder'
df['Sleep Disorder'] = df['Sleep Disorder'].fillna('None')

plt.figure(figsize=(5,4))
df['Sleep Disorder'].value_counts().plot(kind='bar')
plt.title("Distribution of Sleep Disorder")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()

```

```python
# Correlatie heatmap (alle numerieke kolommen) zonder person ID
num_cols = [col for col in df.select_dtypes(include=['int64', 'float64']).columns if col != 'Person ID']

# Correlatie heatmap zonder Person ID
plt.figure(figsize=(12,10))
sns.heatmap(df[num_cols].corr(), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()


# Pairplot van een selectie belangrijke variabelen
cols_for_pairplot = [
    'Sleep Duration',
    'Quality of Sleep',
    'Stress Level',
    'Physical Activity Level',
    'Heart Rate',
    'Daily Steps'
]

# Filter alleen kolommen die echt bestaan (voor de zekerheid)
cols_for_pairplot = [c for c in cols_for_pairplot if c in df.columns]

sns.pairplot(df[cols_for_pairplot])
plt.show()
```

```python


# Outlier analyse met boxplots op numerieke kolommen
plt.figure(figsize=(15, 10))
for i, col in enumerate(num_cols, 1):
    plt.subplot(4, 4, i)
    sns.boxplot(x=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()

# (Optioneel) Outliers trimmen m.b.v. IQR-methode
# Pas dit alleen toe als jullie dit in de tekst ook toelichten
for col in num_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    df = df[(df[col] >= lower) & (df[col] <= upper)]

print("Dataset shape na outlier filtering:", df.shape)
```

```python
# Categorische kolommen bepalen
cat_cols = df.select_dtypes(include=['object']).columns
print("Categorische kolommen:", list(cat_cols))

# We droppen 'Person ID' omdat dit een identificatieveld is en geen feature
if 'Person ID' in df.columns:
    df = df.drop(columns=['Person ID'])

# We willen categorische features encoden, maar de target 'Sleep Disorder' apart behandelen
feature_cat_cols = [c for c in cat_cols if c in df.columns and c != 'Sleep Disorder']

print("Categorische feature-kolommen die we gaan encoden:", feature_cat_cols)

# One-hot encoding voor categorische feature-kolommen
df_encoded = pd.get_dummies(df, columns=feature_cat_cols, drop_first=True)

# Target 'Sleep Disorder' label-encoden naar numerieke waarden
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df_encoded['Sleep Disorder'] = le.fit_transform(df_encoded['Sleep Disorder'].astype(str))

df_encoded.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Features (X) en target (y) scheiden
X = df_encoded.drop('Sleep Disorder', axis=1)
y = df_encoded['Sleep Disorder']

print("Feature shape:", X.shape)
print("Target shape:", y.shape)

# Schalen van numerieke features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test-split (80/20) met vaste random_state voor reproduceerbaarheid
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y  # zorgt voor gelijke klassenverdeling in train en test
)

print("Train shape:", X_train.shape, y_train.shape)
print("Test shape:", X_test.shape, y_test.shape)
```

## Waarom deze dataset geschikt is voor ons ML-vraagstuk

Voordat we een machine learning model kunnen bouwen, moeten we bepalen of de dataset
geschikt is om onze onderzoeksvraag te beantwoorden:

**"In hoeverre kunnen we op basis van lifestyle- en gezondheidsfactoren voorspellen of iemand een slaapstoornis heeft?"**

Om dit te beoordelen hebben we de dataset eerst grondig geanalyseerd. Hieronder lichten wij toe waarom deze dataset geschikt is voor dit vraagstuk.


### 1. De dataset bevat een duidelijke target variabele
De kolom **Sleep Disorder** bevat drie mogelijke uitkomsten:
- Insomnia  
- Sleep Apnea  
- None (geen slaapstoornis)

Hiermee is de dataset direct toepasbaar voor **supervised classification**, wat precies aansluit op ons vraagstuk.



### 2. De dataset bevat relevante en logische features
De dataset bevat meerdere variabelen die wetenschappelijk onderbouwd invloed hebben op slaapkwaliteit en slaapstoornissen, zoals:

- **Sleep Duration**  
- **Quality of Sleep**  
- **Stress Level**  
- **Physical Activity Level**  
- **BMI Category**  
- **Heart Rate**  
- **Daily Steps**

Uit de EDA blijkt dat deze variabelen duidelijke variatie vertonen en betekenisvolle relaties hebben met elkaar. Dit zijn goede voorspellers voor een ML-model.

### 3. De datakwaliteit is hoog
Uit onze analyses (info(), describe(), missing values) blijkt dat:

- Er zijn **geen ontbrekende waarden** in de belangrijke variabelen  
- Datatypes zijn correct (numeriek, categorisch)  
- Waarden liggen in **realistische medische ranges**  
- De dataset bevat **voldoende variatie** voor modeltraining

Dit betekent dat de data betrouwbaar en volledig genoeg is voor machine learning.



### 4. De EDA toont betekenisvolle patronen
Onze correlatieanalyse en pairplots laten zien dat:

- **Sleep Duration** en **Quality of Sleep** sterk positief correleren  
- **Stress Level** negatief correleert met slaapkwaliteit en slaapduur  
- **Heart Rate** duidelijk stijgt bij hogere stress  
- **Daily Steps** sterk samenhangt met **Physical Activity Level**

Deze patronen passen bij bestaande medische inzichten en bevestigen dat de dataset inhoudelijk klopt.



### 5. Outlieranalyse bevestigt stabiliteit
Behalve enkele outliers bij Daily Steps, liggen alle variabelen binnen normale grenzen.
Outliers zijn beheersbaar via IQR-filtering, waardoor de dataset na opschoning stabiel is voor modeltraining.



### 6. De dataset is groot genoeg
Met **1528 rijen** hebben we:

- voldoende datapoints om modellen te trainen  
- voldoende diversiteit tussen personen  
- een betrouwbare basis voor evaluatiemetrics  



### 7. Geschikt voor meerdere ML-modellen
Omdat de dataset zowel lineaire als niet-lineaire relaties laat zien:

- Logistic Regression is geschikt als baseline  
- Random Forest, Decision Tree en KNN kunnen non-lineaire patronen oppakken  

Dit maakt de dataset flexibel en rijk genoeg voor diverse modelvergelijkingen.



## Conclusie
Op basis van de datakwaliteit, variatie, relevante features, duidelijke targetvariabele en de gevonden patronen in de EDA concluderen wij dat deze dataset **geschikt en betrouwbaar** is om het onderzoeksvraagstuk te beantwoorden en machine learning-modellen te trainen voor het voorspellen van slaapstoornissen.
